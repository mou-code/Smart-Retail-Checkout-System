```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("test2.jpg", cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blur = cv2.GaussianBlur(gray, (9, 9), 0)

# Apply Canny edge detection
edges = cv2.Canny(blur, 90, 10)

# Use contour detection
dilation_kernel = np.ones((3, 3), np.uint8)
edges_dilated = cv2.dilate(edges, dilation_kernel, iterations=1)
erosion_kernel = np.ones((3, 3), np.uint8)
edges_eroded = cv2.erode(edges_dilated, erosion_kernel, iterations=1)
plt.imshow(cv2.cvtColor(edges_eroded, cv2.COLOR_BGR2RGB))
plt.show()
contours, _ = cv2.findContours(edges_eroded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


# Draw rectangles around detected objects
rectangle_positions = []
print(f"contours len:  {len(contours)}")
for contour in contours:
    # Ignore small contours (adjust the threshold as needed)
    if cv2.contourArea(contour) > 500:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        rectangle_positions.append((x, y, x + w, y + h))

print(rectangle_positions)

# Display the result
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
##dynamic thresh
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("test2.jpg", cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blur = cv2.GaussianBlur(gray, (9, 9), 0)

# Adaptive Gaussian Thresholding
edges = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.show()
# Morphological Closing
closing_kernel = np.ones((5, 5), np.uint8)
edges_closed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, closing_kernel, iterations=1)
plt.imshow(cv2.cvtColor(edges_closed, cv2.COLOR_BGR2RGB))
plt.show()
# Use contour detection on the morphologically closed image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(contours)

# Draw rectangles around detected objects
rectangle_positions = []
print(f"contours len: {len(contours)}")
for contour in contours:
    # Adjust the contour area threshold dynamically
    dynamic_threshold = max(0.01 * cv2.arcLength(contour, False), 100)
    
    if cv2.contourArea(contour) >dynamic_threshold:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        rectangle_positions.append((x, y, x + w, y + h))

print(rectangle_positions)

# Display the result
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
